```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("penguins.csv")
df.head()
```

```python
df= df.dropna()

df.isna().sum()
```

```python
df = df.drop(df[df['sex'] == "."].index)
df['sex'].unique()
```

#### The culmen is "the upper ridge of a bird's beak" 

```python
sns.scatterplot(x="culmen_length_mm", y="culmen_depth_mm", data=df, hue="species")
plt.title("Culmenl Length vs Culmen Depth")
```

```python
sns.boxplot(x = "species", y = "flipper_length_mm", data =df)
plt.title("Flipper Length for 3 Penguin Species",  color="purple")
```

```python
sns.pairplot(df, hue = "species", height=3)
```
