```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("penguins.csv")
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df= df.dropna()

df.isna().sum()
```

```python
df.shape
```

```python
df['sex'].unique()
```

```python
df = df.drop(df[df['sex'] == "."].index)
df['sex'].unique()
```

```python
df.info()
```

species: penguin species (Chinstrap, Ad√©lie, or Gentoo)
<br>culmen_length_mm: culmen length (mm)
<br>culmen_depth_mm: culmen depth (mm)
<br>flipper_length_mm: flipper length (mm)
<br>body_mass_g: body mass (g)
<br>island: island name (Dream, Torgersen, or Biscoe) in the Palmer Archipelago (Antarctica)
<br>sex: penguin sex

```python
sns.histplot(x = "body_mass_g", data = df, hue = "species")
plt.title("Body Mass By Species")
plt.show()
```

```python
sns.catplot(x= "sex", y="flipper_length_mm", hue ="species",
            kind= "bar", data = df)
plt.title("Flipper Length by Species and Sex")
plt.show()
```

```python
sns.heatmap(df.corr(), cmap = "Blues")
```
