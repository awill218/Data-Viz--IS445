```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("store_data.csv", encoding="windows-1252")
df.head()
```

```python
fig = plt.figure(figsize=(10,10))

#Create bar plot
plt.bar(df['Region'], df['Profit'], color = ['purple'])

#axis labels and title
plt.xlabel('Region')
plt.ylabel('Profit')
plt.title('Profit by Region')
```
